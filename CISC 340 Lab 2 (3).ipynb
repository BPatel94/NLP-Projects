The fundamental constraint of the map coloring problem is no two adjacent country may be assigned the same color.
Therefore, when the algorithm attempts to assign a color to a country, it must check all neighboring regions listed in the adjacency structure.Backtracking is triggered if a neighboring country has already been assigned a color. That color is the same as the color currently being attempted. When this condition occurs, the current partial solution is invalid, and the algorithm must undo the assignment and try a different color.

Backtracking Proceeds:
When the constraint is violated. The current country’s color assignment is removed. The algorithm tries the next color. If all colors fail. The algorithm backtracks to the previous country. Changes its color. Continues the search. This prevents exploring invalid solution branches.

```python
class Backtrack:
    @staticmethod
    def colorMap(map_graph):
        colors = ["Red", "Blue", "Black", "Orange"] # List of allowed colors

        assignment = {}  # Stores final assignments
       
        def is_valid(country, color):  # Helper function to check if color is valid
            for neighbor in map_graph[country]:
                if neighbor in assignment and assignment[neighbor] == color:
                    return False
            return True
        def backtrack(countries): # Backtracking recursive function
            if not countries:    # Base case: all countries colored
                return True
            country = countries[0]  # Pick next country
            for color in colors:  # Try each color
                if is_valid(country, color):
                    assignment[country] = color
                    if backtrack(countries[1:]): # Recursing on remaining countries
                        return True
                    del assignment[country]    # Backtrack step
            return False
        backtrack(list(map_graph.keys()))  # Start backtracking
        return [[country, assignment[country]] for country in assignment]  # Convert result to array of arrays
```

```python
def queens(n): # Static method queens take input n = number of queens and board size (n × n)
    board = [-1] * n  # -1 means no queen placed yet
    def is_safe(row, col):  # Function check queen safe or not at position (row, col)
        for prev_row in range(row): # Loop all previous rows where queens already placed
            if board[prev_row] == col: # If another queen in same column false
                return False
            if abs(board[prev_row] - col) == abs(prev_row - row): #Check diagonal attack
                return False
        return True # If no attack found good position
    def backtrack(row): # Define recursive function called backtrack
        if row == n: # Base condition
            return True
        for col in range(n): # Try queen position one by one
            if is_safe(row, col): # Check placing queen safe or not
                board[row] = col # Store column value inside board list
                if backtrack(row + 1): # Recursively call next row
                    return True
                board[row] = -1 # Backtracking step and remove queen from current row
        return False # If all columns tried but no safe position
    backtrack(0)
    return [(row, board[row]) for row in range(n)]
```
